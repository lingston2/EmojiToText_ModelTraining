```python
import pandas as pd
import csv
import numpy as np
import math
```

```python
# PD_Emoji = pd.read_csv("/content/drive/MyDrive/FinalProjectIntroML/Dataset_Emoji.csv")
PD_Emoji = pd.read_csv("/content/drive/MyDrive/FinalProjectIntroML/DSEmojiToSentence.csv")
PD_Emoji.dropna(inplace=True)
PD_Emoji.reset_index(inplace=True, drop=True)
# PD_Emoji.drop(columns="Number", inplace=True)
PD_Emoji.reset_index(inplace=True, drop=True)
PD_Emoji = PD_Emoji.sample(frac=1, ignore_index=True)
```

```python
PD_Emoji
```

```python
Emoji_SR = PD_Emoji["String_representation"].tolist()
Emoji_SR_Cleaned = []
for entries in Emoji_SR:
  NewString = entries.replace("::", ", ")
  NewString = NewString.replace(":", "")
  # NewString = "Please create a sentence from these 3 aspects: " + NewString #For Emoji->Text
  NewString = "Please respond the following text as if in a chat using only one sentence: " + NewString #For Emoji->Response
  Emoji_SR_Cleaned.append(NewString)
```

```python
Emoji_SR_Cleaned
```

```python
Emoji_S_Result = PD_Emoji["Sentence"].tolist()
Emoji_S_Resulted = []
for entries in Emoji_S_Result:
  NewString = "Result: " + entries
  Emoji_S_Resulted.append(NewString)
```

```python
Emoji_S_Resulted[0]
```

```python
# Combination
Emoji_Combined = []
CounterEntries = 0
for entries in Emoji_SR_Cleaned:
  NewFinalString = entries + ". " + Emoji_S_Resulted[CounterEntries]
  Emoji_Combined.append(NewFinalString)
  CounterEntries = CounterEntries+1
```

```python
Emoji_Combined
```

```python
FinalVersionEmoji = pd.DataFrame(Emoji_Combined, columns=['prompt_and_question'])
```

```python
FinalVersionEmoji
```

```python
def SplitData(DataList):
  ##70% Training
  ##20% Validation
  ##10% Test
  LenDataList = len(DataList)
  LenTraining = math.floor(LenDataList*70/100)
  LenTest = math.floor(LenDataList*10/100)
  LenValidation = LenDataList - LenTraining - LenTest
  print(LenTraining, end=" ")
  print(LenValidation, end=" ")
  print(LenTest, end="")
  TrainingList = DataList[:LenTraining]
  ValidationList = DataList[LenTraining:LenTraining+LenValidation]
  TestList = DataList[LenTraining+LenValidation:]
  return TrainingList, ValidationList, TestList

FinTraining, FinValidation, FinTest = SplitData(FinalVersionEmoji)
```

```python
FinTraining.to_csv("training.csv", index=False)
FinValidation.to_csv("validation.csv", index=False)
FinTest.to_csv("test.csv", index=False)
```
